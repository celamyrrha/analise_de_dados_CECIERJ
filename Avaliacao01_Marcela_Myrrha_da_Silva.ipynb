**(0,5 ponto)** 2.c) Você sabe o que é palíndromo? Palíndromo é uma palavra, frase ou número que permanece igual quando lidos de trás para frente. Por exemplo, _arara_, _Ana_, _luz azul_, etc. Crie uma função que cheque se a palavra de entrada é um palíndromo. 






```python
# Questao 2
# 2.c) 

# Funcao
def palindromo(palavra):
  palavra2 = palavra[::-1]
  if palavra2 == palavra:
    print("É palíndromo!")
  else:
    print("Não é palíndromo :(")
  

# Teste
palindromo("casa")
palindromo("asa")
```

**(0,5 ponto)** 2.e) Suponha que uma empresa cobra taxa de 5,5% de multa para atrasos em pagamentos. Crie uma função que calcule o valor atualizado de uma dívida. 

> Dica: o montante (ou total) da dívida $m$ é igual a $v \times \frac{(100+p)}{100}$ ou $v + v \times \frac{p}{100}$. 

```python
# Questao 2
# 2.e) 

# Funcao
def multa(divida):
  valor_atual = divida + (divida*(5.5/100))

  return print(valor_atual)
# Teste
multa(1000)
```

**(0,5 ponto)** 3.d) O banco está codificado. Vamos criar variáveis que _traduzam_ os códigos. Crie no banco `credito` as seguintes variáveis:

`sexo2`: `sexo = 1, sexo2 = Masculino` ou  `sexo = 2, sexo2 = Feminino` 

`escolaridade2`: `escolaridade <= 2, escolaridade2 = Ensino superior ou maior` ou `escolaridade >= 3, escolaridade2 = Ensino medio ou menor` 

`status_civil2`: `status_civil = 1, status_civil2 = Casado(a)` ou  `status_civil = 2, status_civil2 = Solteiro(a)` ou `status_civil = 3, status_civil2 = Outros`

`pagamento2`: `pagamento < 0, pagamento2 = Antecipado` ou `pagamento = 0, pagamento2 = Em dia` ou `pagamento > 0, pagamento2 = Atrasado`

```python
# Questao 3
# 3.d) 
#Criação da coluna sexo2
credito.loc[credito["sexo"] == 1, "sexo2"] = "Masculino"
credito.loc[credito["sexo"] == 2, "sexo2"] = "Feminino"

#Criação da coluna escolaridade2
credito.loc[credito["escolaridade"] <= 2, "escolaridade2"] = "Ensino superior ou maior"
credito.loc[credito["escolaridade"] >= 3, "escolaridade2"] = "Ensino medio ou menor"

#Criação da coluna status_civil2
credito.loc[credito["status_civil"] == 1, "status_civil2"] = "Casado(a)"
credito.loc[credito["status_civil"] == 2, "status_civil2"] = "Solteiro(a)"
credito.loc[credito["status_civil"] == 3, "status_civil2"] = "Outros"

#Criação da coluna pagamento2
credito.loc[credito["pagamento"] < 0, "pagamento2"] = "Antecipado"
credito.loc[credito["pagamento"] == 0, "pagamento2"] = "Em dia"
credito.loc[credito["pagamento"] > 0, "pagamento2"] = "Atrasado"

credito
```
