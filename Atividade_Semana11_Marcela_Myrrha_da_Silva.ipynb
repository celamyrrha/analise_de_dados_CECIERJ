###Questão 1

O Índice de Desenvolvimento Humano (IDH) é um índice estatístico composto de expectativa de vida, educação (média de anos de escolaridade completados e anos esperados de escolaridade ao entrar no sistema educacional) e indicadores de renda per capita, que é usado para classificar os países em quatro níveis do desenvolvimento humano. Um país obtém um nível mais alto de IDH quando a expectativa de vida é mais alta, o nível de educação é mais alto e a renda nacional bruta per capita é mais alta.

Abaixo, temos dados do IDH para alguns países. Crie um gráfico que melhor se adeque aos dados conforme as boas práticas. Argumente sobre a escolha desse gráfico.

```python
# Questao 1

# Carregando biblioteca
from matplotlib import pyplot as plt

# Dados
pais = ["Suíça", "EUA", "Portugal", "Brasil", "Indonésia"]
idh = [0.962, 0.921, 0.866, 0.754, 0.705]

# Grafico
plt.bar(pais, idh)
plt.title("Comparação de IDH")

# Exibe o grafico
plt.show()

# Por que você escolheu esse tipo de gráfico?
#> Acho que esse tipo de gráfico gera uma boa visualização dos dados, facilitando a comparação do índice.

# Observando o gráfico, em termos de IDH, o Brasil está mais próximo da Indonésia
# ou de Portugal?
#> O Brasil está mais próximo da Indonésia, onde vemos uma pequena diferença de altura nas colunas.
```

###Questão 2

Esta é uma lista da população das unidades federativas brasileiras segundo estimativas de 25 de dezembro de 2022, com base nos dados da prévia do censo de 2022 do Instituto Brasileiro de Geografia e Estatística (IBGE). 

Por curiosidade, o Estado com a maior população do Brasil é São Paulo. Em contraponto, Roraima possue a menor população.

Temos interesse em ver como a população se distribue no país. Crie um gráfico que melhor se adeque aos dados conforme as boas práticas. Argumente sobre a escolha desse gráfico.

```python
# Questao 2

# Dados
populacao = [41262199, 19597330, 15989929, 14016906, 10444526, 10693929, 8796448, 
          8452381, 7581051, 6248436, 6003788, 6574789, 3514952, 3766528, 3483985,
          3035122, 3168027, 3118360, 3120494, 2570160, 2449024, 2068017, 1562409,
          1383445, 733559, 669526, 450479] 

# Grafico
populacao.sort()
for i in range(len(populacao)):
  populacao[i] = populacao[i]/1000000

plt.hist(populacao, bins=5)

plt.title('População dos Estados Brasileiros')
plt.xlabel('Número de cidades por faixa')
plt.ylabel('Milhões de habitantes')

# Exibe o grafico
plt.show()

# Por que você escolheu esse tipo de gráfico?
#> Esse é gráfico indicado para quando se tem somente uma variável.
```

Crie um (único) gráfico que relacione as `vendas` com a `temperatura` média registrada naquele mês. Argumente sobre a escolha desse gráfico.
